# Box Plots

The box plot was developed by John Wilder Tukey in 1969.  He was a mathemtician who believd that too much emphasis was placed on confirmatory data analysis rather than exploratory analysis.  Hence the box plot was developed as a tool for exploratory analysis providing a simple graphical representation of a probability distribution.

The box plot is a visual method of representing the distribution of data based on the spread and the interquartile range (IQR) or what is also called the five-number summary.  The interquartile range is the middle fifty percent of the data set and is bounded by the 25th and 75th percentile, the median marks the centre of the dataset as the 50th percentile and the maximum and minimum describe the upper and lower bounds of the data.

Although box plots lack the statistical detail that other visualisations may have they are convenient and take up less space, so it is easy to compare distributions in multiple datasets in the same visualisation for example the image below.

![PDF](images/multi_boxplots.png)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixing random state for reproducibility
np.random.seed(19810826)


#create random sample set using a normal distribution with mean 500, std dev 100
data = np.random.normal(500,100,1000)
```

```python
plt.boxplot(data, vert=False)
plt.show()
```

The box plot above displays a box plot.  This consists of the following components:
- The band in the centre of the box is the second quartile
- The left and right extents of the box mark the first and third quartiles
- The lines extend from the box are the whiskers
- The circles at the ends of the lines show outlying data points

### The Quartiles

The quartiles are defined by the median value of the data set.  The first quartile is then calculated as the median of the lower half and the third quartile is the median of the upper half.  The second quartile is the same as the calculated median of the entire dataset.  The term percentile is also used when describing quartiles, in statistics the term percentile indictaes the upper bound of where a group of observations will fall for example 95% of observations will fall within the 95th percentile.

- The first quartile is the median of the lower half of the data set.  It is also known as the 25th percentile.

- The second quartile is the median of the entire dataset.  It is also known as the 50th percentile.

- The thrid quartile is the median of the upper half of the data set.  It is also known as the 75th quartile.

### The Whiskers

The common variation of a box plot displays whiskers that extend 1.5*IQR beyond the first and third quartiles.  In a normal distribution this extends to below 3 standard deviations of the data.  Datapoints beyond this inerval are considered outliers.

The default in matplolib for whiskers is 1.5*IQR howver these can  be changed in the arguments for the function. The factor can be entered or percentile e.g 5th and 95th or if range is entered then the whiskers will extend to the full range of the data set.

```python
#generate random numbers in a normal distribution
s = np.random.randn(1000)

print("IQR*1.5")
plt.boxplot(s, vert=False)
plt.show()
print("range")
plt.boxplot(s, vert=False, whis='range')
plt.show()
```

### Notches

Notches can also be used on the boxes around the medians to offer a guide on the significance of different medians.  The size of the notches are proportional to the IQR and inversely proportional to the square root of the sample size.

```python
print("notched")
plt.boxplot(s, vert=False, notch=True)
plt.show()
```

### Alternatives

A simple alternative to the boxplot is the histogram.  This provides a more holistic view of the data set but does not display explicit values such as the quartiles or medians in the visual element.  The generally have to be infered or annotated.  The other disadvantage of a histogram over a box plot is the plotting multiple distibutions on the same axis.  This can be done with a histogram but more than a couple the visual will become cluttered.

A more complex version of the boxplot is a violin plot.  This is similar to the boxplot however each side shows a rotated kernel density plot providing a more context to the probability distribution.

```python
#example taken from https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Seaborn_Cheat_Sheet.pdf
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns

#dataset taken from seaborn library
titanic = sns.load_dataset("titanic")

sns.violinplot(x="age",y="sex", hue="survived",data=titanic)
plt.show()
```
