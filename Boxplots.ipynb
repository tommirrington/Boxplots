The box plot is a visual method of representing the distribution of data based on the spread and the interquartile range or what is also called the five-number summary.  The interquartile range is the middle fifty percent of the data set and is bounded by the 25th and 75th percentile, the median marks the centre of the dataset as the 50th percentile and the maximum and minimum describe the upper and lower bounds of the data.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixing random state for reproducibility
np.random.seed(19810826)


#create random sample set using a normal distribution with mean 500, std dev 100
data = np.random.normal(500,100,1000)
data
```

```python
plt.boxplot(data, vert=False)
```

### The Quartiles

The quartiles are defined by the median value of the data set.  The first quartile is then calculated as the median of the lower half and the third quartile is the median of the upper half.  The second quartile is the same as the calculated median of entire dataset.
